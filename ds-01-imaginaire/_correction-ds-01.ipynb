## Question 5)

La fonction $u\mapsto D_N$(u)est une fonction paire, continue sur $\mathbb{R}$ (la fonction est évidemment continue en tout  $u$ où $\sin(\pi u)\neq 0$ c'est-à-dire $u$ non entier, et est prolongée par continuité aux points entiers). Sur $[-1/2,1/2]$, elle admet un lobe principal situé entre les deux abscisses où $D_N$ s'annule, c'est-à-dire entre 
$-\frac{1}{2N+1}$ et $\frac{1}{2N+1}$. La valeur en $u=0$ est $2N+1$ et la valeur en $\pm 1/2$ est $(-1)^{N}$. Le graphe de la fonction est 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
v= np.arange(-0.5,0.5,0.001)
v=v+0.00001;   ### selon vous, pourquoi ai-je ajouté ce 10e-5?
N=10
D10 = (np.sin(np.pi*(2*N+1)*v))/(np.sin(np.pi*v))

plt.figure(6,figsize = (10, 10))
plt.plot(v,D10,label=r'$D_N(t)$ pour $N=10$')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
