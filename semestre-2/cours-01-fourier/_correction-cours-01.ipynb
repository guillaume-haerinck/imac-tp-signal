# Un exemple de corrigé d'un exercice proposé en séance

**Enoncé** : il s'agit de comprendre le rôle des  paramètres $\alpha>0$ et $\sigma>0$ sur la fonction 
$$e(t)=t^{\alpha}\exp(-\sigma t)$$
que nous avons appelé *fonction enveloppe*. 

```python
import numpy as np
import matplotlib.pyplot as plt 

def enveloppe(t , alpha, sigma): 
    "<t> est un vecteur; <alpha> : terme t^alpha, <sigma> : atténuation"
    out = (t**alpha)*(np.exp(-sigma*t));  ## observez bien l'utilisation des opérations terme-à-terme
    return out

###  pour sortir du corps de la fonction : revenez à la ligne et supprimez l'indentation
pas = 1/100
t=np.arange(0,10,pas)  ## un vecteur de valeurs équi-réparties entre 0 et 10
a=1; s=1
e1=enveloppe(t,a,s)
a=3; s=1
e2=enveloppe(t,a,s)
a=3; s=2
e3=enveloppe(t,a,s)
plt.figure(1)
plt.clf  ## effacer la figure le cas échéant
plt.plot(t,e1,color='red',label=r'$\alpha=1 \ , \ \sigma=1$')
plt.plot(t,e2,color='blue',label=r'$\alpha=3 \ , \ \sigma=1$')
plt.plot(t,e3,color='green',label=r'$\alpha=3 \ , \ \sigma=2$')
plt.grid()
plt.legend(loc='upper right')

```

On remarque qu'il est assez malaisé de comparer les différentes courbes à cause d'un problème d'échelle. Nous allons donc normaliser la fonction $t \mapsto e(t)$ pour que son maximum soit égal à $1$ quel que soit le jeu de paramètres $(\alpha,\sigma)$.
  A cette fin, nous allons calculer la position du maximum en étudiant les variations de la fonction $t \mapsto e(t)$. La dérivée s'écrit :
  $$e'(t) = \left( -\sigma t + \alpha   \right) t^{\alpha - 1} \exp(-\sigma t).$$
  Cette fonction s'annule en $t_0=\frac{\alpha}{\sigma}$ et est positive sur $[0,t_0]$, négative sinon. 
  Le maximum de la fonction est donc en $t_0$ et vaut 
  $$e_\max = \left( \frac{\alpha}{\sigma} \right)^\alpha \exp (-\alpha).$$
  
  Ce calcul montre aussi que pour $\alpha < 1$, la pente de la courbe est infinie (montée très rapide) alors que cette pente est nulle pour $\alpha>1$ (montée lente) et égale à $1$ pour $\alpha=1$. 
  
  Travaillons maintenant avec la "nouvelle" fonction enveloppe dans sa version normalisée : 
  


```python
def enveloppe_n(t , alpha, sigma): 
    "<t> est un vecteur; <alpha> : terme t^alpha, <sigma> : atténuation"
    out = (t**alpha)*(np.exp(-sigma*t));  
    valeur_max = ((alpha/sigma)**alpha)*np.exp(-alpha)
    out = out / valeur_max
    return out


a=1; s=1
e1=enveloppe_n(t,a,s)
a=3; s=1
e2=enveloppe_n(t,a,s)
a=3; s=2
e3=enveloppe_n(t,a,s)
plt.figure(2)
plt.clf  ## effacer la figure le cas échéant
plt.plot(t,e1,color='red',label=r'$\alpha=1 \ , \ \sigma=1$')
plt.plot(t,e2,color='blue',label=r'$\alpha=3 \ , \ \sigma=1$')
plt.plot(t,e3,color='green',label=r'$\alpha=3 \ , \ \sigma=2$')
plt.legend(loc='upper right')
plt.grid()
plt.title('fonction enveloppe normalisée')
```

Si l'on souhaite une enveloppe qui "ressemble" (dans nos rêves) à l'enveloppe d'un son de piano, on poura choisir par exemple $\alpha=0.5$ : 


```python
a=0.5; s=1
e1=enveloppe_n(t,a,s)
e2=enveloppe_n(t,a,2)

plt.figure(3)
plt.clf  ## effacer la figure le cas échéant
plt.plot(t,e1,color='red',label=r'$\alpha=0.5 \ , \ \sigma=1$')
plt.plot(t,e2,color='blue',label=r'$\alpha=0.5 \ , \ \sigma=2$')
plt.legend(loc='upper right')
plt.grid()
plt.title('fonction enveloppe normalisée')
```
